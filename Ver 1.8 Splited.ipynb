```python
# Hàm crawl data
tech = ['Vật Vờ Studio',
       'Duy Thẩm',
       'Tony Phùng Studio',
       'Tân Một Cú',
       'AnhEm TV',
       'Vinh Xô',
       'Tinh Tế',
       'Relab',
       'Khôi Ngọng',
       'Duy Luân Dễ Thương',
       'Công Nghệ Lõi',
       'MobileCity',
       'Tuấn Ngọc đây!',
       'Lee Tùn Zân - Review Và Tặng Lại',
       'ĐAM MÊ CÔNG NGHỆ']

beauty = ['Ha Linh official',
        'Chloe Nguyen',
        'Changmakeup',
        'Góc của Rư',
        'Mai Vân Trang',
        'Call me Duy',
        'Luna Đào Official',
        'Nguyen Thuc Thuy Tien',
        'An Phương',
        'Q U I N',
        '✨ Gau Zoan ✨',
        'Rosie Pham',
        'Blingbabi',
        '1m88',
        # 'Kylie Jenner' --> Not Vietnamese
        ]

food = ['Ông Anh thích nấu ăn',
        'Bếp Trưởng Review',
        'Ninh Tito',
        'Khoai lang thang',
        'Cô Ba Bình Dương',
        'Sapa TV',
        'Dũng béo ẩm thực phố phường',
        'VNT Food & Travel',
        'ĐI ĐÂU ĂN GÌ ?',
        'Ẩm thực mẹ làm',
        'Hanoi food',
        'Max McFarlin',
        'Thằng Mập Food',
        'Hôm nay ăn gì?',
        'Ok Con Dê']  

def main(lst):
    df = pd.DataFrame(columns = ['Channel', 'Subscribers', 'Subscribers for the last 30 days', '% change', 'Video views for the last 30 days', '% change'])
    browser = webdriver.Edge(executable_path="msedgedriver.exe")
    browser.get("https://socialblade.com/")
    sleep(1)

    for i in range(len(lst)):
        search = browser.find_element(By.ID, 'SearchInput')
        sleep(1)
        search.send_keys(lst[i])
        search.send_keys(Keys.ENTER)

        
        subscriber = browser.find_element(By.XPATH, "/html/body/div[13]/div/div[3]/div[3]/span[2]")  
                                #         /html/body/div[12]/div/div[3]/div[3]/span[2]      
        sleep(1)

        subscriber30 = browser.find_element(By.XPATH, "/html/body/div[18]/div[1]/div[1]/div[3]/div[1]/p[1]")
                                                    /html/body/div[17]/div[1]/div[1]/div[3]/div[1]/p[1]
        s30 = subscriber30.text.split(' ',-1)
        s30num = s30[0]
        if s30num == '--':
            s30num = None
            s30per = None
        else:
            s30p = s30[1]
            strn = browser.find_element(By.XPATH, '/html/body/div[18]/div[1]/div[1]/div[3]/div[1]/p[1]/sup/span/i')  
                                                /html/body/div[17]/div[1]/div[1]/div[3]/div[1]/p[1]/sup/span/i     
            strn = strn.get_attribute('class')
            if strn == 'fa fa-caret-down':
                st = '-'
            else:
                st = '+'
            s30per = st + s30p
        sleep(1)

        vids30 = browser.find_element(By.XPATH, "/html/body/div[18]/div[1]/div[1]/div[3]/div[3]/p[1]")
                                                 /html/body/div[17]/div[1]/div[1]/div[3]/div[3]/p[1]
        v30 = vids30.text.split(' ',-1)
        v30num = v30[0]
        if v30num == '--':
            v30num = None
            v30p = None
            vt = None
        else:
            v30p = v30[1]
            vtrn = browser.find_element(By.XPATH, '/html/body/div[18]/div[1]/div[1]/div[3]/div[3]/p[1]/sup/span/i')
                                                /html/body/div[17]/div[1]/div[1]/div[3]/div[3]/p[1]/sup/span/i
            vtrn = vtrn.get_attribute('class')
            if vtrn == 'fa fa-caret-down':
                vt = '-'
            else:
                vt = '+'
            v30per = vt + v30p
        sleep(1)

        df.loc[len(df)] = [lst[i], subscriber.text, s30num, s30per, v30num, v30per]
    
    browser.close()
    return df
```

```python
# Dọn dẹp dữ liệu crawl về/ Chuyển từ kiểu string về float 
def convert_to_numberString(df):
    df['Subscribers'] = df['Subscribers'].str.replace('K','000')
    df['Subscribers'] = df['Subscribers'].str.replace('M','0000')
    df['Subscribers'] = df['Subscribers'].str.replace('.','')

    df['Subscribers for the last 30 days'] = df['Subscribers for the last 30 days'].str.replace('K','000')
    df['Subscribers for the last 30 days'] = df['Subscribers for the last 30 days'].str.replace('M','0000')
    df['Subscribers for the last 30 days'] = df['Subscribers for the last 30 days'].str.replace('.','')

    df['Video views for the last 30 days'] = df['Video views for the last 30 days'].str.replace('K','')
    df['Video views for the last 30 days'] = df['Video views for the last 30 days'].str.replace('M','000')
    df['Video views for the last 30 days'] = df['Video views for the last 30 days'].str.replace('.','')


    # Change "%change" into number 
    df['% change'] = df['% change'].str.replace('%','')
    df['% change.1'] = df['% change.1'].str.replace('%','')
    df["% change"] = pd.to_numeric(df["% change"])
    df["% change.1"] = pd.to_numeric(df["% change.1"])
    # df["% change.1"] = pd.to_numeric(df["% change.1"], downcast="float")

    #Change type to float
    df["Subscribers"] = pd.to_numeric(df["Subscribers"], downcast="float")
    # df['Subscribers'] = df['Subscribers'].astype(float)
    df['Subscribers for the last 30 days'] = df['Subscribers for the last 30 days'].astype(float)
    df['Video views for the last 30 days'] = df['Video views for the last 30 days'].astype(float)

convert_to_numberString(df)
df=df.drop(['Unnamed: 0'], axis=1)
df = df.fillna(df.mean()) # Doesn't work yet cause still string
```

```python
# Tạo historgram và chuyển về thang histogram. VD: 2 tr lượt sub -> 4đ
des_table = df.describe()
maxSub =  des_table["Subscribers"].values[7]
maxSub30 = des_table["Subscribers for the last 30 days"].values[7]
maxView30 = des_table["Video views for the last 30 days"].values[7]

def histo_maker(series,maxOf): 
    class_width = round((series.values[7] - series.values[3])/5)
    class_width2 = class_width*2
    class_width3 = class_width*3
    class_width4 = class_width*4
    return class_width,class_width2,class_width3,class_width4, maxOf

def converFunc (series, maxOf):
    for i in range (len(series)):
        if series.values[i] < class_width:
            series.values[i] =2.0
        elif series.values[i]>= class_width and series.values[i]< class_width2:
            series.values[i] =2.5
        elif series.values[i]>= class_width2 and series.values[i]< class_width3:
            series.values[i] =3.0
        elif series.values[i]>= class_width3 and series.values[i]< class_width4:
            series.values[i] =3.6
        elif series.values[i]>= class_width4 and series.values[i] <= maxOf:
            series.values[i] =4.0 

class_width,class_width2,class_width3,class_width4, maxOf = histo_maker(des_table["Subscribers"],maxSub)
converFunc(df['Subscribers'], maxOf)
class_width,class_width2,class_width3,class_width4, maxOf = histo_maker(des_table["Subscribers for the last 30 days"], maxSub30)
converFunc(df['Subscribers for the last 30 days'], maxOf)
class_width,class_width2,class_width3,class_width4, maxOf = histo_maker(des_table["Video views for the last 30 days"], maxView30)
converFunc(df['Video views for the last 30 days'], maxOf)
```
